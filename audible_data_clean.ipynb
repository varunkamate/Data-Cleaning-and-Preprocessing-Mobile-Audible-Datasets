```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("audible_uncleaned.csv")
```

```python
df
```

```python
df["name"]=df["name"].str.replace("[^A-Za-z&0-9,:]","",regex=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
import re

def convert_to_minutes(time_str):
    hours = minutes = 0
    h_match = re.search(r'(\d+)\s*hr', time_str)
    m_match = re.search(r'(\d+)\s*min', time_str)

    if h_match:
        hours = int(h_match.group(1))
    if m_match:
        minutes = int(m_match.group(1))

    return hours * 60 + minutes

df["time"]=df["time"].apply(convert_to_minutes)
```

```python
df
```

```python
df['releasedate'] = pd.to_datetime(df['releasedate'], format='%d-%m-%y')
df['releasedate'] = df['releasedate'].dt.strftime('%Y-%m-%d')
```

```python
df
```

```python
df['num_ratings'] = df['stars'].str.extract(r'(\d+)\s*ratings').astype(float)
```

```python
df["stars"]=df["stars"].apply(lambda x:x[0:3])
```

```python
df
```

```python
df["stars"]=df["stars"].str.replace("Not","-",regex=True)
```

```python
df["stars"]=df["stars"].str.replace("[^0-9.]","",regex=True)
```

```python
df
```

```python
df.to_csv("audible_cleaned",index=False)
```
