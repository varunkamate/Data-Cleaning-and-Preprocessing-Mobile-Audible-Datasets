```python
#for to fill out the null values to the column we are using simple imputer
#because all columns which contain the null values are same data type

from sklearn.impute import SimpleImputer

df.select_dtypes("object").columns
```

```python
a=pd.DataFrame(n,columns=df.select_dtypes("object").columns)
```

```python
df["Name"]=a["Name"]
```

Feature extraction opportunities

From battery: extract capacity (mAh) and charging power (W).

From storage: separate RAM and internal storage.

From display: extract screen size (inches), resolution, refresh rate.

From camera: extract rear and front camera megapixels.

From processor: extract chipset and clock speed.

```python
df["battery_capacity"] = df["battery"].str.extract(r'(\d+)\s*mAh')[0]+"mAh"
```

```python
df["charging_capacity"]=df["battery"].str.extract(r'(\d+)\s*W')[0]+"W"
```

```python
df.drop(columns=["battery"],inplace=True)
```

```python
df
```

```python
df["ram"]=df["storage"].str.extract(r'(\d+)\s*GB RAM')[0]+"GB RAM"
df["internal_storage"]=df["storage"].str.extract(r'(\d+)\s*GB')[0]+"GB"
```

```python
#From display: extract screen size (inches), resolution, refresh rate.
df["screen_size"]=df["display"].str.extract(r'(\d+)\s*inches')[0]+'inches'
df["resolution"]=df["display"].str.extract(r'(\d+\s*×\s*\d+\s*px|\d+\s*x\s*\d+)')[0]+'px'
df["refresh_rate"]=df["display"].str.extract(r'(\d+)\s*Hz')[0]+'Hz'
```

```python
#From camera: extract rear and front camera megapixels.
df['rear_camera']=df['camera'].str.extract(r'^(.*?)&')
df["front_camera"]=df['camera'].str.extract(r'&\s*(.*)')
```

```python
#From processor: extract chipset and clock speed.
df["chipset"] = df["processor"].str.extract(r'^(.*?),')
df["clock_speed"] = df["processor"].str.extract(r'(\d+\.?\d*\s*GHz)')
```
